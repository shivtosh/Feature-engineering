```python
y=d3['Pass/Fail']
x=d3.drop(['Pass/Fail'],axis=1)
```

```python
from scipy.stats import zscore
dScaled=x.apply(zscore)
dScaled.head()
```

```python
import numpy as np
pca = PCA().fit(dScaled)
plt.grid()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
#around 200 can describe almost 98% of variance
pca = PCA(n_components=200)
pca.fit(dScaled)
Xpca = pca.transform(dScaled)
```

```python
plt.bar(list(range(1,201)),pca.explained_variance_ratio_,alpha=0.5, align='center')
plt.ylabel('Variation explained')
plt.xlabel('eigen Value')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test1, y_train, y_test1 = train_test_split(Xpca, y, test_size=0.40, random_state=7)
X_valid,X_test,y_valid,y_test=train_test_split(X_test1,y_test1, test_size=0.50, random_state=8)
```

```python
print("Before UpSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before UpSampling, counts of label '-1': {} \n".format(sum(y_train==-1)))
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
```

```python
y_predict = model1.predict(X_train)
model1_score = model1.score(X_train, y_train)
print(model1_score)
```

```python
y_predict = model1.predict(X_valid)
model1_score = model1.score(X_valid, y_valid)
print(model1_score)
```

```python
y_predict = model1.predict(X_test)
model1_score = model1.score(X_test, y_test)
print(model1_score)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, y_predict))
print(metrics.classification_report(y_test, y_predict))
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1 ,k_neighbors = 5, random_state=1)   #Synthetic Minority Over Sampling Technique
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
```

```python
print("After UpSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After UpSampling, counts of label '-1': {} \n".format(sum(y_train_res==-1)))
```

```python
print('After UpSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After UpSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
```

### Perfomance analysis of model after upsampling

```python
y_predict1 = model.predict(X_train)
model_score = model.score(X_train_res, y_train_res)
print(model_score)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, y_predict1))
print(metrics.classification_report(y_test, y_predict1))
```
